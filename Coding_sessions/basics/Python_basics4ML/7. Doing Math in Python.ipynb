# <font face="gotham" color="purple"> Vectors </font>

````python
import numpy as np
````

Every vector in the vector space has a length except $0$-vector, and the terminology of length is **norm**, the general formula of norm of a vector in $\mathbb{R}^k$is 

$$
\|u\|= \sqrt{x_1^2+x_2^2+...+x_k^2}
$$

As an example, let's plot vectors $(4, 7)$ and $(8, 6)$ in $\mathbb{R}^2$.

There is a NumPy function for computing norms: ```np.linalg.norm()```.

````python
a = np.array([4, 7])
b = np.array([8, 6])
a_norm = np.linalg.norm(a)
b_norm = np.linalg.norm(b)
print('Norm of a and b are {0:.4f}, {1} respectively.'.format(a_norm, b_norm))
````

# Vector Addition, Subtraction and Scalar Multiplication.

The vector addition is element-wise operation, if we have vectors $\mathbf{u}$ and $\mathbf{v}$ addition:

$$\mathbf{u}+\mathbf{v}=\left[\begin{array}{c}
u_{1} \\
u_{2} \\
\vdots \\
u_{n}
\end{array}\right]+\left[\begin{array}{c}
v_{1} \\
v_{2} \\
\vdots \\
v_{n}
\end{array}\right]=\left[\begin{array}{c}
u_{1}+v_{1} \\
u_{2}+v_{2} \\
\vdots \\
u_{n}+v_{n}
\end{array}\right]$$

And subtraction:
$$\mathbf{u}+(-\mathbf{v})=\left[\begin{array}{c}
u_{1} \\
u_{2} \\
\vdots \\
u_{n}
\end{array}\right]+\left[\begin{array}{c}
-v_{1} \\
-v_{2} \\
\vdots \\
-v_{n}
\end{array}\right]=\left[\begin{array}{c}
u_{1}-v_{1} \\
u_{2}-v_{2} \\
\vdots \\
u_{n}-v_{n}
\end{array}\right]$$

And the scalar operations:
$$c \mathbf{u}=c\left[\begin{array}{c}
u_{1} \\
u_{2} \\
\vdots \\
u_{n}
\end{array}\right]=\left[\begin{array}{c}
c u_{1} \\
c u_{2} \\
\vdots \\
c u_{n}
\end{array}\right]$$

# <font face="gotham" color="purple"> Matrix Operations</font>

````python
import sympy as sy
````

````python
# You can do  !pip install ->"Library"
````

Matrix _addition_ operations are straightforward:
1. $A+ B= B+ A$
2. $(A+B)+ C=A+(B+C)$
3. $c(A+B)=cA+cB$
4. $(c+d)A=cA+c{D}$
5. $c(dA)=(cd)A$
6. $A+{0}=A$, where ${0}$ is the zero matrix
7. For any $A$, there exists an $- A$, such that $ A+(- A)=0$.

The matrix _multiplication_ properties are:
1. $ A({BC})=({AB}) C$
2. $c({AB})=(cA)B=A(cB)$
3. $A(B+ C)={AB}+{AC}$
4. $(B+C)A={BA}+{CA}$

Note that we need to differentiate two kinds of multiplication, **Hadamard multiplication** (element-wise multiplication) and **matrix multiplication**: 

````python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
````

````python
A*B # this is Hadamard elementwise product
````

````python
A@B # this is matrix product
````

````python
A = sy.Matrix([[3, 4], [7, 8]])
B = sy.Matrix([[5, 3], [2, 1]])
A*B
B*A
````

# <font face="gotham" color="purple"> Transpose of Matrices </font>

Matrix $A_{n\times m}$ and its transpose is 

````python
A = np.array([[1, 2, 3], [4, 5, 6]])
A.T # transpose
````

````python
A = sy.Matrix([[1, 2, 3], [4, 5, 6]]); A
A.transpose()
````

The properties of transpose are
1. $(A^T)^T$
2. $(A+B)^T=A^T+B^T$
3. $(cA)^T=cA^T$
4. $(AB)^T=B^TA^T$
