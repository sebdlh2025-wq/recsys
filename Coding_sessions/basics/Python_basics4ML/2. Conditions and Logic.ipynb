# Loops

```python
#For loop
for name in names:
    # print(name)
    if name == 'elon musk':
        print('Hello musk\n')
        #next value in loop
        continue
    elif name == 'yuri gagarin':
        print('hey yuori')
    elif name == 'nel armstorng':
        print('Hello Nel')
    elif name == 'bereket':
        print('wow')
        break
    else:
        print('name not part of condition')
    print('Pleased to have you here \n')
```

```python
launched_suttles = 0
total_shuttles =10


while True:
    # String formatting
    print('Shuttle {} launched'.format(launched_suttles+1))
    launched_suttles +=1 #increment by1

    if launched_suttles == total_shuttles:
        print('All shuttels lauched')
        break
```

```python
astronauts = {
            "names": "Elon musk",
            "suit size": "lage",
            "alergies": "peanust"
}
```

```python
for key in astronauts.keys():
    print(key)
```

```python
#Loop through keys and values at the same time
for key, value in astronauts.items():
    print(key)
    print(value, '\n')
```
