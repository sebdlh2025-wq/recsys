# Functions and String formatting

```python
#Defining a function in Python
def launcpad_welcome():
    print('Hello')
    print(1+2)
    print(3+4)
```

```python
launcpad_welcome()
```

```python
for idx in range(10):
    launcpad_welcome()
```

```python
names = ['neil armstrong', 'buzz aldrin', 'sally ride', 'yuri gagarin','elon musk']
```

```python
#Positional argument and string formatting

'''The .format() method formats the specified value(s) and
    insert them inside the string's placeholder. 
    The placeholder is defined using curly brackets: {}.'''

def custom_welcome(name):
    print('Welcome {}'.format(name))
```

```python
custom_welcome('nel armstrong')
```

```python
for name in names:
    custom_welcome(name)
```

```python
print('Welcome {} {}'.format('Leonard Hofstadter','Sheldon Cooper'))
```

```python
#Multiple Positional argument and string formatting
def custom_welcome_toship(name, spaceship):
    print('Welcome {} to the {}'.format(name,spaceship))
```

```python
custom_welcome_toship('Leonard','Galactic_one')
```

```python
space_ship = ['Galactic 1', 'Galactic 2', 'Abisinya 1', 'Sheldonleeship', 'Abay ship' ]
```

```python
space_ship[3]
```

```python
'''Python's enumerate() lets you write Pythonic for loops
when you need a count and the value from an iterable. '''


for idx, name in enumerate(names):
    print(idx,name)
```

```python
for idx, name in enumerate(names):
    ship = space_ship[idx]
    custom_welcome_toship(name,ship)
```

```python
#Keyword argument
def space_suit(color = 'Blue'):
    print('Your space suit is {}'.format(color))
```

```python
space_suit()
```

```python
space_suit(color = 'Red')
```

```python
def space_suit_welcome(name, space_ship, color ='Blue'):
    print('welcome {} to the {}, your space suit is {}'.format(name,space_ship,color))
```

```python
for idx, name in enumerate(names):
    ship = space_ship[idx]
    space_suit_welcome(name, ship, color = 'Red' )
```

```python
space_suit_welcome('Leonard','Galctic',)
```

```python
def space_suit_welcome(name, space_ship, color ='Blue', allergies = 'None'):
    print('welcome {} to the {}, your space suit is {}, you have {} allergies'.format(name,space_ship,color, allergies))
```

```python
#Keyword argument position does NOT matter
space_suit_welcome('Howard','Galctic',allergies ='peanut', color = 'White')
space_suit_welcome('Howard','Galctic', color = 'White', allergies ='peanut')
```

```python
#Set up a return statement
def space_suit_welcome_with_return(name, space_ship, color ='Blue', allergies = 'None'):
    return 'welcome {} to the {}, your space suit is {}, you have {} allergies'.format(name,space_ship,color, allergies)
```

```python
space_suit_welcome_with_return('Howard','Galctic')
```

```python
#returning a value and storing it in a variable
Welcome = space_suit_welcome_with_return('Alex','Galctic')
```

```python
print(Welcome)
```
