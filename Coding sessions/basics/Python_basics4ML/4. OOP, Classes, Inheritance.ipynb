# OOP, Classes, Inheritance

# 1. Classes

Classes are objects with defiend properties and methods.

```python
class Person():
    #The method that runs as soon as you create a class --initialise
    def __init__(self, name, age, color): #a function attached to a class
        #Create some attrtibutes for person class
        self.name = name
        self.age = age
        self.color = color
        
    # Date of birth method
    def Year_of_birth(self): #pass self so that this fuction has access to name,age and color
        return 2022 - self.age
    
    #Projected age
    def projected_age(self, years=5):
        return self.age+years
```

```python
new_person = Person('Elon Musk', 50, 'blue')
```

```python
#accessing a class attribute
new_person.name
```

```python
# Run a method
new_person.Year_of_birth()
```

```python
#Run a method with a keyworkd argument
new_person.projected_age(years = 10)
```

# Inheritance

```python
# Take all the attributes and Methods of one class and asign them to another classs (Parent to child class)
#Parent = person
#Child = Astronaut


class Astronaut(Person):
    def __init__(self,name, age, color, mission_length_in_month):
        #This is what gives us inheritance = SUPER METHOD
        super().__init__(name,age,color)
        self.mission_length_in_month = mission_length_in_month
    #Method for calculating age on return
    def age_on_return(self):
        return self.projected_age(years = int(self.mission_length_in_month/12))
        
```

```python
new_astronaut = Astronaut('Howard', 29, 'white', 48 )
```

```python
#Access parent like attribute
new_astronaut.color
```

```python
new_astronaut.mission_length_in_month
```

```python
#Access method
new_astronaut.Year_of_birth()
```

```python
#Run a method
new_astronaut.age_on_return()
```
